**Task 4. - Data Cleaning**

```python
data.isnull().sum()

#dropping all missing values rows
data.dropna(inplace =True)

data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with duplicate rown
data.duplicated().sum()

# viewing duplicate records
# data[data.duplicated()]

# deleting duplicate records
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# tyoe casting
# changing data types
data.dtypes

data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

### EDA ###
**Task 5. - Data Analysis**

***Univariate Analysis***

```python
# identifying outliers in price

df = data[data['price'] < 1500]

sns.boxplot(data=df, x='price')
```

```python
#Price distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
#Availability distribution

plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
# avg price per bed
df.groupby(by= 'neighbourhood_group')['price per bed'].mean()
```

```python
df.groupby(by= 'neighbourhood_group')['price'].mean()
```

**Feature Engineeing**

```python
# ['price per bed']

df['price per bed']= df['price']/df['beds']
df.head()
```

**Bi variable Analysis**
one variable dependency in another variable

```python
df.columns
```

```python
# Price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price per bed', hue='room_type')
plt.title('Price dependency on neighbourhood')
```

```python
# number of reviews and price relationship

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title('Locality and Review Dependency')
plt.show()
```
